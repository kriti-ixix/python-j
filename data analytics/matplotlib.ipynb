```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.plot(x,y,'r',linestyle='dashed',linewidth=2,marker="*",markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2d-diagram')
plt.show()
```

# bar plot

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2)
plt.title('bar graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# functional methods

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.subplot(1,2,2)
plt.plot(x,y,'r')

plt.subplot(1,2,1)
plt.plot(x,y,'g')
```

```python
plt.subplot(2,2,4)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'g')

plt.subplot(2,2,3)
plt.plot(x,y,'y')

plt.subplot(2,2,1)
plt.plot(x,y,'b')
```

# object oriented methods

```python
fig=plt.figure()

axes= fig.add_axes([0.1,0.2,0.9,0.9])
#add_axes(left,bottom,width,height)
axes.plot(x,y)
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.2,0.9,0.9])
axes1.plot(x,y,'r',label='red')

axes2=fig.add_axes([0.15,0.75,0.3,0.3])
axes2.plot(x,y,'b',label='blue')
axes1.legend(loc=4)
axes2.legend()

```
