```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, '#FF00FF')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('My Graph')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
plt.subplot(2, 2, 1)
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3)
plt.subplot(2, 2, 4)

plt.tight_layout()
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```
